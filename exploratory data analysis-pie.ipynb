```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("modified_healthcare_dataset.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Gender','Blood Type','Medical Condition','Insurance Provider','Admission Type','Test Results']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Comparison Of {dt}")
    plt.ylabel("")
    plt.show()
```
